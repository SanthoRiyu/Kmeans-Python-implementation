### Kmeans - 2 coordinate , k=2

```python
from scipy.spatial import distance
from tabulate import tabulate
import pandas as pd

# Points
x=[2,2,8,5,7,6,1,4]
y=[10,5,4,8,5,4,2,9]


#Clusters
m =[2,5]
n =[10,8]


def average_cord(x,ind):
    av1,av2 = 0,0
    for ind1 in ind:
        av1+=x[ind1][0]
        av2+=x[ind1][1]
    return (av1/len(ind), av2/len(ind))

coords1 = list(zip(x,y))
centroids = list(zip(m,n))

ind = ['P'+str(i) for i in range(1,len(x)+1)]
head = ['C'+str(i) for i in range(1,len(m)+1)]

ind_df = [i for i in range(1,len(x)+1)]
head_df = [i for i in range(1,len(m)+1)]

def distance_table(coords1, centroids):
    df= distance.cdist(coords1, centroids, 'euclidean')
    df = pd.DataFrame(df,index = ind_df, columns=head_df)
    df['Cluster'] = df.idxmin(axis=1)
    head.append('Cluster')
    print(tabulate(df.values,headers=head,showindex=ind,tablefmt='fancy_grid'))

    ind1 = [x-1 for x in list(df[df.Cluster==1].index)]
    ind2 = [x-1 for x in list(df[df.Cluster==2].index)]

    centroids = []
    for var1 in [ind1, ind2]:
        centroids.append(average_cord(list(zip(x,y)), var1))

    for ii in range(1,len(centroids)+1):
        print(f'Cluster C{ii} is {list(df[df.Cluster==ii].index)}')
    print(f'New_centroids - {centroids}\n')
    head.remove('Cluster')
    return centroids, df.Cluster.values
        
centroids, cluster_form = distance_table(coords1, centroids)

while True:
    cluster_form1 = cluster_form
    centroids, cluster_form = distance_table(coords1, centroids)
    if (cluster_form1 == cluster_form).all():
        break
```
